```python
def mergeanomalies(label):
    if label in [
        "DDoS",
        "DoS Hulk",
        "DoS GoldenEye",
        "DoS Slowhttptest",
        "DoS slowloris"
    ]:
        return "1"
    elif label=="PortScan":
        return "2"

    elif label in [
        "FTP-Patator",
        "SSH-Patator",
        "Web Attack – Brute Force"
    ]:
        return "3"

    elif label=="BENIGN":
        return "0"
    else:
        return "4"
```

```python
df[" Label"] = df[" Label"].apply(mergeanomalies)
```

```python

from sklearn.preprocessing import LabelEncoder

import numpy as np

X = df.drop(columns=[" Label"])
y = df[" Label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    stratify=y,
    test_size=0.2,
    random_state=42
)
```

```python
# Replace inf with NaN
X_train = X_train.replace([np.inf, -np.inf], np.nan)

# Drop rows with NaN and reset index
X_train = X_train.dropna().reset_index(drop=True)

# Reset labels to match
y_train = pd.Series(y_train).reset_index(drop=True)

# Trim y to same length (important)
y_train = y_train.iloc[:len(X_train)]
```

```python
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

classes = np.unique(y_train)

weights = compute_class_weight(
    class_weight="balanced",
    classes=classes,
    y=y_train
)

class_weights = dict(zip(classes, weights))

print(class_weights)
```

```python
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler

counts = pd.Series(y_train).value_counts()

TARGET = 300000   # change based on RAM

sampling_strategy = {
    cls: min(TARGET, count)
    for cls, count in counts.items()
}

rus = RandomUnderSampler(
    sampling_strategy=sampling_strategy,
    random_state=42
)

X_small, y_small = rus.fit_resample(X_train, y_train)

print("After undersampling:")
print(pd.Series(y_small).value_counts())
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(
    random_state=42,
    k_neighbors=5
)

X_res, y_res = smote.fit_resample(X_small, y_small)

print("After SMOTE:")
print(pd.Series(y_res).value_counts())
```

```python
# Combine X_res and y_res into a single dataframe
df_smote = pd.DataFrame(X_res)
df_smote[' Label'] = y_res.values

# Save the SMOTE-resampled dataset
output_path = r"C:\Users\sudee\OneDrive\Documents\GitHub\MINOR_2-NETWORK-ANAMOLY-DETECTION\DATASET\SMOTE_resampled_dataset.csv"
df_smote.to_csv(output_path, index=False)

print(f"Dataset saved to: {output_path}")
print(f"Shape: {df_smote.shape}")
print(f"\nClass distribution:\n{df_smote[' Label'].value_counts()}")
```
